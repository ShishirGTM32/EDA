```python
df3 = df2[['race_day','race_name','race_length', 'athlete_id', 'athlete_gender','athlete_age', 'athlete_performance','athlete_average_speed', 'race_no_of_finishers','year']]
```

```python
sns.histplot(df3['race_length'])
plt
```

```python
sns.lmplot(data=df3, x = 'athlete_age', y = 'athlete_average_speed')

plt.title('Athlete Average Speed Vs Athlete Age', color = 'red')
plt.xlabel('Athlete Age')
plt.ylabel('Athlete Averag Speed')
plt.show
```

```python
sns.lmplot(data=df3, x = 'athlete_age', y = 'athlete_average_speed',hue = 'athlete_gender')

plt.title('Athlete Average Speed Vs Athlete Age & Gender', color = 'red')
plt.xlabel('Athlete Age')
plt.ylabel('Athlete Average Speed')
plt.show
```
